```python
import cv2
from keras.models import load_model
import numpy as np
import random
import time
model = load_model('keras_model.h5')
cap = cv2.VideoCapture(0)   # (cap) Object that controls the input of the camera
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32) # This is a matrix - dimension (1, 2, 3,Then RGB). One image per iteration consisting on 224 pixels wide and high

started = False
next_round = True
countdown = False
counter = 0
elapsed = 0
global str
chosen = False
c = 1
score = 0
games = 0
CompletedRound = True
t_0 = time.time
message = ''
message_shown = False
kaylanWins = False
computerWins = False
kaylanScore = 0
computerScore = 0

class Players():
    def __init__(self, won_game, score=0, games=0):
        self.score = score
        self.games = games
        self.won_game = won_game

    def add_to_score(self, point):
        self.score = self.score + point
        self.a_win() # delete
        return
    
    def who_wins(self):
        self.won_game += 1
        # if self.won_game % 1 == 0:
        self.add_to_score(1)

    def set_to_zero(self):
        self.score = self.score - self.score
        return
    
    def a_win(self): # delete
        self.score == 3
            

kaylan = Players(0)
computer = Players(0)
    


def get_user_choice(prediction):
    if prediction[0][0] > 0.6: # Assigns variables to the predictions
        player_choice = 'Rock'
    elif prediction[0][1] > 0.6:
        player_choice = 'Paper'
    elif prediction[0][2] > 0.6:
        player_choice = 'Scissors'
    else:
        player_choice = 'Nothing'
    return player_choice
            
            
            

def compare_options(computer_choice, player_choice):
    if player_choice == computer_choice:
        print('Computer:', computer_choice)
        print('Player:', player_choice)
        message = 'DRAW'
    elif computer_choice == 'Rock' and player_choice == 'Paper':
        print('Computer:', computer_choice)
        print('Player:', player_choice)
        message = 'YOU WIN'
        
    elif computer_choice == 'Rock' and player_choice == 'Scissors':
        print('Computer', computer_choice)
        print('Player', player_choice)
        message = 'YOU LOSE'
        
    elif computer_choice == 'Scissors' and player_choice == 'Rock':
        print('Computer:', computer_choice)
        print('Player', player_choice)
        message = 'YOU WIN'
        
    elif computer_choice == 'Scissors' and player_choice == 'Paper':
        print('Computer:', computer_choice)
        print('Player', player_choice)
        message = 'YOU LOSE'
        
    elif computer_choice == 'Paper' and player_choice == 'Rock':
        print('Computer', computer_choice)
        print('Player', player_choice)
        message = 'YOU LOSE'
        
    elif computer_choice == 'Paper' and player_choice == 'Scissors':
        print('Computer', computer_choice)
        print('Player', player_choice)
        message = 'YOU WIN'
        
    else:
        print('Computer', computer_choice)
        print('Player', player_choice)
        message = 'Waiting'
    return message

            
while True: 
    ret, frame = cap.read() # A that moment it will capture the image in a matrix or an array
    resized_frame = cv2.resize(frame, (224, 224), interpolation = cv2.INTER_AREA)
    image_np = np.array(resized_frame)
    normalized_image = (image_np.astype(np.float32) / 127.0) - 1 # Normalize the image
    data[0] = normalized_image
    prediction = model.predict(data)

    if not started:
        message = 'Press a to start'
        if cv2.waitKey(33) == ord('a'):
            # if not started:
                counter = time.time()
                started = True
                countdown = True
    
    if started:
        elapsed = 7 - (time.time() - counter) # countdown from 7 seconds
        if countdown:
            message = f'Show your hand in {int(elapsed)} seconds'
            if elapsed <= 0:
                countdown = False
        
        
        elif elapsed <=0: # if the countdown is <= 0 
            countdown = False # stops the countdown
            
            if c > 0:
                player_choice = get_user_choice(prediction)
                computer_choice = random.choice(['Rock', 'Paper', 'Scissors'])
                c = c - 1
                message = f'You showed {player_choice}, and the computer showed {computer_choice}.'
            
           
                

            elif elapsed <= -11: # change to '-9'
                message = 'Press n to play the next round'
                if cv2.waitKey(32) == ord('n'):
                    started = False
                    elapsed = 0     
                    c = 1
                    CompletedRound = True
                    kaylanWins = False
                    computerWins = False
                    
            elif kaylanScore == 3: # kaylan.add_to_score(3) == 3 
                message = "You've won the game - press 'q' to quit!"
                cv2.waitKey(32) == ord('q')    

            elif computerScore == 3:
                message = "You lose, computer wins!! Press 'q' to quit."
                cv2.waitKey(32) == ord('q')
        
            elif elapsed <= -6:
                message = compare_options(computer_choice, player_choice)
                if CompletedRound:
                    if message == 'YOU WIN':
                        CompletedRound = False
                        kaylanWins = True
                        kaylanScore = kaylanScore + 1
                        
                        
                        
                    elif message == 'YOU LOSE':
                        CompletedRound = False
                        computerWins =True
                        computerScore = computerScore + 1
                        
                        
                    else:
                        message = 'DRAW' or 'Waiting'
                        CompletedRound = False
                
    #if kaylan.a_win():
    #    message = "You've won the game!"       
    #elif computer.a_win():
    #    message = 'Computer wins the game'


         
            
        
            
    cv2.putText(frame, message, (30, 30), cv2.FONT_HERSHEY_DUPLEX, 0.5, (0, 0, 255), 2)
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break
            
# After the loop release the cap object
cap.release()
    # Destroy all the windows
cv2.destroyAllWindows()
# When you press 'Run' it will show the probability of each category one after another (rock, paper, scissors, nothing)
#Need to take the highest value and translate that into the classes that we have
```
